```python
import pandas as pd
import xlrd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
total_data = pd.read_csv("clean_data_v4.csv")
```

```python
total_data.plot(kind='box',y='Total Charges',x='APR Severity of Illness Code')
```

```python
df = pd.DataFrame({total_data['APR Severity of Illness Code'],'col1': total_data['Total Charges']})

fig, ax = plt.subplots(figsize=(10,  10))
df.boxplot(['col1'], 'APR Severity of Illness Code', ax)
```

```python
total_data.boxplot('Total Charges','APR Severity of Illness Code')
```

```python
ax = total_data.boxplot('Total Charges','APR Severity of Illness Code')
ax.set_ylim(0,total_data['Total Charges'].quantile(0.95))
ax.set_title('')
ax.set_ylabel('Total Charges')
```

```python
f,(ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,5))
total_data.boxplot('Total Charges','APR Severity of Illness Code',ax=ax1)
ax1.set_ylim(0,total_data['Total Charges'].quantile(0.95))
total_data.boxplot('Total Charges','Age Group',ax=ax2)
ax1.set_title('')
ax2.set_title("")
ax1.set_ylabel('Total Charges')
f.suptitle("")
```

```python
f, ((ax1, ax2), (ax3, ax4),(ax5, ax6)) = plt.subplots(3, 2, sharey=True,figsize=(10,5))
total_data.boxplot('Total Charges','APR Severity of Illness Code',ax=ax1,showfliers=False)
ax1.set_ylim(0,total_data['Total Charges'].quantile(0.95))
total_data.boxplot('Total Charges','Age Group',ax=ax2,showfliers=False)
total_data.boxplot('Total Charges','Gender',ax=ax3,showfliers=False)
total_data.boxplot('Total Charges','Race',ax=ax4,showfliers=False)
ax3.set_ylim(0,total_data['Total Charges'].quantile(0.95))
total_data.boxplot('Total Charges','Emergency Department Indicator',ax=ax5,showfliers=False)
total_data.boxplot('Total Charges','Day_group',ax=ax6,showfliers=False)
ax5.set_ylim(0,total_data['Total Charges'].quantile(0.95))
ax1.set_title('')
ax2.set_title('')
ax3.set_title('')
ax4.set_title('')
ax5.set_title('')
ax6.set_title('')
ax1.set_ylabel('Total Charges')
ax1.set_xlabel('APR Severity of Illness Code')
ax3.set_ylabel('Total Charges')
ax5.set_ylabel('Total Charges')
f.suptitle("")
```
